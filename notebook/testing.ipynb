```python
fraud_sample = base[base.fraud_bool == 1].sample(n = 10_000)

non_fraud_sample = base[base.fraud_bool == 0].sample(n = 10_000)

selected_sample = pd.concat([fraud_sample, non_fraud_sample])

selected_sample = selected_sample.reset_index(drop=True)
```

```python
selected_sample.fraud_bool.value_counts()
```

```python
from pycaret.classification import setup, compare_models, pull, save_model
target = selected_sample.fraud_bool
setup(selected_sample, target= target)
```

```python
setup_df = pull()
```

```python
setup_df
```

```python
best_model = compare_models()
compare_df = pull()
```

```python
compare_df
```

```python
save_model(best_model, 'best_model')
```
